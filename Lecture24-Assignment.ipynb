```python
import user_agents
```

```python
def get_User_info(p_UserAgent):
    ua = user_agents.parse(p_UserAgent)
    return pd.Series((ua.os.family,ua.browser.family,ua.device.family))
```

```python
vist_df[['os_family','browser_family','device_family']]=vist_df['UserAgent'].apply(get_User_info)
```

```python
vist_df['os_family'].value_counts()
```

```python
def get_ip_info(ip):
    try:
        response = DbIpCity.get(ip, api_key='free')
        return pd.Series((response.country,response.region,response.city,response.latitude,response.longitude))
    except:
        return 'Error'
```

```python
vist_df100[['country','region','city','latitude','longitude']] = vist_df100['IP'].apply(get_ip_info)
```
