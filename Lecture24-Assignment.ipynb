```python
vist_df = pd.read_csv('visitor-interests.csv')
```

```python
vist_df
```

```python
import user_agents
```

```python
def get_User_info(p_UserAgent):
    ua = user_agents.parse(p_UserAgent)
    return pd.Series((ua.os.family,ua.browser.family,ua.device.family))
```

```python
vist_df[['os_family','browser_family','device_family']]=vist_df['UserAgent'].apply(get_User_info)
```

```python
vist_df
```

```python
vist_df['os_family'].value_counts()
```

```python
vist_df['browser_family'].value_counts()
```

```python
vist_df['device_family'].value_counts()
```

```python
def Concat_string(x):
#     0 - First Column
#     1 - Second Column
     return str(x[0])+' ' + str(x[1])
```

```python
vist_df.info()
```

```python
vist_df['OS-Device'] = vist_df[['os_family','device_family']].apply(Concat_string,axis=1)
```
